```python
#break statement
#find the first even number in a list

numbers=[1,2,3,5,8,9,10]

for num in numbers:
    if num%2==0:
        print ("First even number found:",num)
        break
        
```

```python
#search for a specific element in a list

target_element=42
data=[10,20,30,40,50]

for element in data:
    if element==target_element:
        print("Element found")
    else:
        print ("not found")
        break
```
