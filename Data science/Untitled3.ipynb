```python
import pandas as pd

#load the file
df=pd.read_excel('pandas_200_Employees.xlsx')
```

```python
#problem 1:Identify and count the numbr of missing values in each column of the dataset

missing_value_count=df.isnull().sum()
print(missing_value_count)
```

```python
#problem 2: find and remove the duplicate records from the dataset

df_cleaned=df.drop_duplicates()
print("original Rows:",len(df))
print("Rows after removing duplicate:",len(df_cleaned))
```

```python
#problem3:normalize the 'Age' columns using the min-max normalization technique.
#X_normalozed=(x-x_min)/(x_max-X_min)
min_age=df['Age'].min()
max_age=df['Age'].max()

df['Normalized_Age']=(df['Age']-min_age)/(max_age-min_age)
print(df[['Age','Normalized_Age']])

```

```python
#problem4:Standardised the 'Salary' column using the standardization technique.
#X_standarized=(x_mean)/standard deviation

mean_salary=df['Salary'].mean()
std_salary=df['Salary'].std()
df['Standardized_Salary']=(df['Salary']-mean_salary)/std_salary
print(df[['Salary','Standardized_Salary']])
```

```python
#problem5:Fill in missing values in the 'Gender' column with the mode value.
mode_gender=df['Gender'].mode()[0]
df['Gender'].fillna(mode_gender,inplace=True)
print(df['Gender'])
```

```python
#problem6: average salary by department
avg_salary_dept=df.groupby("Department")['Salary'].mean()
print(avg_salary_dept)
```

```python
#problem7: Highest paid top 10 employees

top_10_employee=df.sort_values(by="Salary", ascending=False).head(10)
print(top_10_employee)
```

```python
#problem8: Employee hired after 2020
df['HireDate']=pd.to_datetime(df['HireDate'])

employee_after_2020=df[df['HireDate'].dt.year>2020]
employee_after_2020.head()


```

```python
#Gender-Wise Salary Comparison

gender_salary=df.groupby('Gender')['Salary'].mean()
gender_salary
df.groupby('Gender')['Salary'].agg(['min','max','mean'])
```
