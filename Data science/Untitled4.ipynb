```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#what is the distribution of movies released over the year?
data=pd.read_excel('matplotlib_practice_34889.xlsx')


```

```python
data['release_date']=pd.to_datetime(data['release_date'])
data['Release Year']=data['release_date'].dt.year
```

```python
#plot the distribution
plt.figure(figsize=(10,6))
plt.hist(data['Release Year'],bins=30,edgecolor='k')
plt.xlabel('Release Year')
plt.ylabel('Number of movies')
plt.title('Distribution of the movies over released year')
plt.grid(True)
plt.show()
```

```python
#What are the top 10 movie genres in the dataset and thier frequencies
genre_counts=data['genre'].value_counts().head(10)

#plotting the top 10 movie genre

plt.figure(figsize=(10,6))
genre_counts.plot(kind='bar')
plt.xlabel('genre')
plt.ylabel('Number of movie')
plt.title('Top 10 movie genre')
plt.show()
```

```python
#prblm3:show the distribution of movie origins.

#count movie origin

origin_counts=data['origin_ethnicity'].value_counts().head(5)

#plot the distribution 
plt.figure(figsize=(10,6))
origin_counts.plot(kind='pie',autopct='%1.1f%%')
plt.title('Distribution of top 5 movie origin')
plt.show()

```
