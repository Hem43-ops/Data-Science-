```python
#line plot showing the trend of weekly_sales over time for store 
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
data = pd.read_excel('walmart_sales_190.xlsx')

# Convert date column to datetime
data['date'] = pd.to_datetime(data['date'])

# Filter data for store 1
store_1_data = data[data['store'] == 1]

# Create line plot
plt.figure(figsize=(10,6))
sns.lineplot(x='date', y='weekly_sales', data=store_1_data)

plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.title('Weekly Sales Trend for Store 1')
plt.xticks(rotation=90)

plt.show()


```
