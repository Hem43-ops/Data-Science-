```python
import numpy as np
Temperature_data=np.array([
    [20,30,32,31,33],
    [21,22,23,24,25],
    [25,28,30,34,35],
    [30,32,35,37,38]
])

Avg_temperature=np.mean(Temperature_data,axis=1)
print("The average temperature of cities are:",Avg_temperature)
```

```python
#to find the city with highest temperature
Avg_temperature=np.mean(Temperature_data,axis=1)

city_highest_avg_temperature=np.argmax(Avg_temperature)
cities=["mumbai","chennai","kolkata","Delhi"]
print("The highest temperature of cities is:",cities[city_highest_avg_temperature])


```

```python
#calulate the temp anomlaies for each city 
temperature_anomalies=Temperature_data-Avg_temperature[:,np.newaxis]
print(temperature_anomalies)
```
