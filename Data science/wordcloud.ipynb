```python
from wordcloud import wordcloud
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel('movies_dataset_34889.xlsx')
#selecting the specific director
director_name='Director A'
director_data=data[data['Direcotr'].str.contains(director_name,case=False)]

#join all plo tdescriptions into a single string
all_plots=''.join(director_data['plot'])

#create the word cloud for the directors plot 
wordcloud=wordCloud(width=880,height=400,background_colour='white').generate(all_plots)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud,interpolation='bilinear')
plt.title(f'most Common keyowrd in Movie plots for {director_name}')
plt.axis('off')
plt.show()
    
```
